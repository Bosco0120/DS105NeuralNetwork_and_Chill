# Rental Distance to Centre Analysis

```python
import pandas as pd
from plotnine import *

#import csv file
df = pd.read_csv('~/Desktop/DS105L/Project/rental-output.csv')
print(df)
```

```python
summary_table = df.describe()
print(summary_table)
```

```python
from plotnine import *
import pandas as pd
import numpy as np
```

```python
# Take the logarithm of the columns
df['log_distance'] = np.log(df['distance'])
df['log_priceByArea'] = np.log(df['priceByArea'])

# Create a scatter plot using plotnine with a custom color
p = (
    ggplot(df, aes(x='log_distance', y='log_priceByArea'))
    + geom_point()
    + stat_smooth(method='lm', se=False, color='#EB7B7B') # Add line of best fit
    + labs(title='Correlation between Log Distance and Log Price by Area', x='Log Distance', y='Log Price by Area')
    + theme_bw()
)

# Display the plot
print(p)
```

```python
p = (
    ggplot(df, aes(x='distance', y='priceByArea'))
    + geom_point()
    + stat_smooth(method='lm', se=False, color='#EB7B7B') # Add line of best fit
    + labs(title='Correlation between Distance and Price by Area', x='Distance', y='Price by Area')
    + theme_bw()
)

# Display the plot
print(p)
```

```python
# Get a list of the unique province names
provinces = df['province'].unique()

# Create an empty dictionary to hold the dataframes
province_dfs = {}

# Loop over the province names and create a dataframe for each one
for province in provinces:
    province_df = df[df['province'] == province]
    province_dfs[province] = province_df
```

```python
df_barcelona = province_dfs['Barcelona']
df_madrid = province_dfs['Madrid']
df_valencia = province_dfs['València']
df_sevilla = province_dfs['Sevilla']
df_zaragoza = province_dfs['Zaragoza']
df_malaga = province_dfs['Málaga']
df_vizcaya = province_dfs['Vizcaya']
df_granada = province_dfs['Granada']
df_balears = province_dfs['Balears (Illes)']
df_laspalmas = province_dfs['Las Palmas']
```
